```python
import pandas as pd

# read in CSV files
df1 = pd.read_csv('Resources/books_to_movies.csv')
df2 = pd.read_csv('Resources/Goodreads_books_with_genres.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# merge t he DataFrames based on "Title" and "Author"
merged_df = pd.merge(df1,df2[['Title', 'Author', 'isbn13']], on=['Title', 'Author'], how='left')
merged_df
```

```python
merged_df.dtypes
```

```python
merged_df['Movie_Released'] = pd.to_numeric(merged_df['Movie_Released'], errors='coerce').astype('Int64')
merged_df['isbn13'] = pd.to_numeric(merged_df['isbn13'], errors='coerce').astype('Int64')
merged_df
```

```python
merged_df = merged_df.drop(columns=['Unnamed: 0'])
```

```python
merged_df.head()
```

```python
# export merged_df as csv
merged_df.to_csv('Resources/cleaned_books_to_movies.csv', index=False)
```
